```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from scipy.stats import boxcox
```

```python
#copy of the data
```

```python
dfcopy1 = df.copy()
```

```python
Y = dfcopy1['total claim amount']
X = dfcopy1.drop('total claim amount', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_numeric = X.select_dtypes(include=['float64', 'int64'])
X_normalized = pd.DataFrame(scaler.fit_transform(X_numeric), columns=X_numeric.columns)
```

```python
# One-Hot Encoding
categorical_vars = X.select_dtypes(include=['object'])
X_encoded = pd.get_dummies(categorical_vars)
```

```python
# Concatenar las variables numéricas normalizadas y las variables categóricas codificadas
X_final = pd.concat([X_normalized, X_encoded], axis=1)
```

```python
# División en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_final, Y, test_size=0.3, random_state=42)
```

```python
# Entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)
```

```python
# Métricas de evaluación del modelo
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print("Coefficient of Determination (R2):", r2)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
```
